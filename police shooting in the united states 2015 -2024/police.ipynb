# A bar chart (or bar graph)
is a type of data visualization that uses rectangular bars to represent data values. 
Each bar's length or height is proportional to the value it represents, allowing for easy comparison between categories.
 Key Features of a Bar Chart:
Bars: Represent individual categories (e.g., countries) and their values (e.g., population).

Axes:
The horizontal axis (x-axis) typically represents categories (e.g., country names).

The vertical axis (y-axis) represents numerical values (e.g., population size).

Bar Height/Length: Corresponds to the value of the category (the taller or longer the bar, the higher the value).

```python
# Creating Bars
import matplotlib.pyplot as plt
# sample data: Country and populations
countries =['china','india','USA', 'indonesia','parkistan']
population =[1444216107,1393409038,332915073,276361783,225199937]


plt.figure(figsize=(10,6))
bars = plt.bar(countries,population,color ='red')

#Adding title and labels
plt.title('population of countries' ,fontsize =16)
plt.xlabel('countries')
plt.ylabel('populations')

#Adding labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2.,height * 1.01,f'{height:,}', ha='center',va='bottom',fontsize=10)

#Explanation 
# bars: Holds the bar objects created by plt.bar().
# plt.text(): Adds text labels on top of each bar:
# bar.get_x() + bar.get_width()/2. centers the label on the bar.
# height * 1.01 positions the label slightly above the bar.
# f'{height:,}' formats the population with commas for better readability.
# ha='center' and va='bottom': Align the text horizontally in the center and vertically at the bottom of the text.
```

```python
# Country and populations
countries =['china','india','USA', 'indonesia','parkistan']
population =[1444216107,1393409038,332915073,276361783,225199937]

# Create a horizontal bar chart
plt.figure(figsize =(10,6))
bars = plt.barh(countries,population, color = 'blue')


#Adding title and labels
plt.title('population of countries' ,fontsize =16)
plt.xlabel('population', fontsize = 12)
plt.ylabel('Countries', fontsize = 12)

# Adding labels on the right side of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.01,bar.get_y() + bar.get_height()/2.,f'{width:,}', ha='left',va='center',fontsize=10)
```

```python
#Race  and Gender Analysis
#what is the breakdown of incidents by race and gender
race_gender_breakdown = police.groupby(['race','gender']).size()
race_gender_breakdown
```

```python
#Location based analysis
#whats state have the highest number of police shooting
state_counts = police['state'].value_counts()
state_counts
```

```python
#Ensure the date column is in datetime format
police ['date'] = pd.to_datetime(police['date'],errors='coerce')
```

```python
#Time based Analysis
#In which month do most police shooting occur
incident_per_month = police['date'].dt.month.value_counts()
incident_per_month
```

```python
#Most common race
#what is the most common race of the people involved in the police shootings
most_common_race = police['race'].value_counts().idxmax()
race_counts=police['race'].value_counts()
most_common_race,race_counts
```

```python
#How many incidents occurred each year
incidents_per_year = police['date'].dt.year.value_counts()
incidents_per_year
#normalize=True: This option returns the relative frequencies (or proportions) instead of the absolute counts. Instead of showing how many times each value appears,
#it shows what percentage of the total rows that value represents.
```
